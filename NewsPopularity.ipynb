```python
new[' category']=new[' category'].replace([''],'Misc')
```

```python
new['Title'] = new['url'].str.slice(31,)
new['Title']=new["Title"].str[:-1]

new['Date']=new["url"].str.slice(20,)
new['Date']=new["Date"].str[:10]

```

```python
mean= new[' shares'].mean()
Unpop=new[new[' shares']<mean].copy()
Pop=new[new[' shares']>=mean].copy()

Unpop[' popular_mean']='Unpopular'
Pop[' popular_mean']='Popular'

frames = [Unpop, Pop ]
  
result = pd.concat(frames)
display(result)
```

```python
result = result.drop(columns=[' weekday_is_monday', ' weekday_is_tuesday',' weekday_is_wednesday',' weekday_is_thursday',' weekday_is_friday',' weekday_is_saturday',' weekday_is_sunday',' data_channel_is_lifestyle',' data_channel_is_entertainment',' data_channel_is_bus',' data_channel_is_socmed',' data_channel_is_tech',' data_channel_is_world'])

result.to_csv('News.csv')
```
